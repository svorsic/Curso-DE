### **AnÃ¡lisis de las variables con datos nulos**

#### Se exploran los datos y su variedad dentro de las columnas que presentan datos nulos para determinar una acciÃ³n a tomar con las mismas. Se pude notar que una de las principales causantes de los datos nulos es la diferencia entre fallecidos y no fallecidos. Este es un dato relevante, por tanto, se evaluarÃ¡ la manera de que se pueda mantener la distinciÃ³n sin perjudicar la integridad de la base.

```python
dfCol.estado.value_counts()
```

#### La columna estado indica con claridad la distinciÃ³n, por tanto, puede procederse a intervenir las columnas con datos nulos.

```python
dfCol.fecha_muerte.value_counts()
```

#### La columna fecha_muerte presenta demasiados valores nulos. Esto se debe a que la gran mayorÃ­a de los pacientes se ha recuperdado. La fecha resulta relevante para hacer estudios posteriores. Por tanto, se intentarÃ¡ conservar la columna para anÃ¡lisis posteriores. Los datos nulos serÃ¡n reemplazados por N/A (No aplica).

```python
dfCol.fecha_recuperado.value_counts()
```

#### La columna fecha_recuperado tiene un pequeÃ±o nÃºmero de datos nulos. Lo ideal serÃ­a poder conservar la informaciÃ³n aplicando algÃºn mÃ©todo de sustituciÃ³n. Los datos nulos serÃ¡n reemplazados por N/A.

```python
dfCol.tipo_recuperacion.value_counts()
```

#### La columna tipo_recuperacion podrÃ­a ser relevante para posteriores estudios. Se decide sustituir los valores nulos por N/A.

```python
dfCol.nom_grupo_.value_counts()
```

La columna Nom_grupo_ no presenta una distinciÃ³n de relevancia para el estudio, por tanto se decide borrarla.

```python
# eliminacÃ³n de columna nom_grupo_
col_del=['nom_grupo_']
dfCol_2=dfCol.drop(labels=col_del, axis=1)
dfCol_2.columns
```

```python
# SustituciÃ³n de valores en las restantes columnas a preservar.
col_sus= ['fecha_muerte','fecha_recuperado','tipo_recuperacion']
for i in col_sus:
  dfCol_2[i] = dfCol_2[i].fillna('N/A')
```

## Verificando que no hay datos nulos

```python
dfCol_2.info()
```

No se presentan datos nulos, el proceso ha sido exitoso.

## Ajuste de tipos de datos

```python
dfCol_2['id_de_caso'] = dfCol_2['id_de_caso'].astype('int')
dfCol_2['departamento'] = dfCol_2['departamento'].astype('int')
dfCol_2['edad'] = dfCol_2['edad'].astype('int')
dfCol_2['per_etn_'] = dfCol_2['per_etn_'].astype('int')
dfCol_2['ciudad_municipio'] = dfCol_2['ciudad_municipio'].astype('int')
```

```python
dfCol_2['fecha_reporte_web'] = dfCol_2['fecha_reporte_web'].astype('datetime64[ns]')
dfCol_2['fecha_de_notificaci_n'] = dfCol_2['fecha_de_notificaci_n'].astype('datetime64[ns]')
dfCol_2['fecha_inicio_sintomas'] = dfCol_2['fecha_inicio_sintomas'].astype('datetime64[ns]')
dfCol_2['fecha_diagnostico'] = dfCol_2['fecha_diagnostico'].astype('datetime64[ns]')
```

#### VerificaciÃ³n de cambios

```python
dfCol_2.info()
```

## Limpieza de valores redundantes

#### Referencias al municipio y al departamente son redundantes, ya que todos los datos pertenecen al mismo departamento y al mismo municipio, es decir, departamento de Antioquia, municipio de Medellin en Colombia.

```python
dfCol_2.ciudad_municipio_nom.value_counts()
```

```python
dfCol_2.departamento_nom.value_counts()
```

```python
col_Ant=['departamento_nom', 'ciudad_municipio_nom']
dfCol_3=dfCol_2.drop(labels=col_Ant, axis=1)
dfCol_3.columns
```

## Recodificar variables categoricas

```python
alldata=[]
alldata1=[]
variables=[]
for i in dfCol_3.columns:
  if dfCol_3[i].dtype == 'O':
    if len(dfCol_3[i].unique()) ==2:
      print('Variable 2 categorias:',i)
      y= pd.get_dummies(dfCol_3[i])
      alldata.append(y)
      variables.append(i)
    else:
      pass
    print('--------------------')
    if len(dfCol_3[i].unique()) >2:
      print('Variable >2 categorias:',i)
      z= pd.DataFrame()
      z[i]=dfCol_3[i].astype('category').cat.codes
      alldata1.append(z)
      variables.append(i)
  if dfCol_3[i].dtype == 'float64':
    dfCol_3[i]=(dfCol_3[i]-dfCol_3[i].mean())/(dfCol_3[i].std())
data_y=pd.concat(alldata, axis=1)
data_z=pd.concat(alldata1, axis=1)
dfCol_Ok= pd.concat([data_y,data_z,dfCol_3],axis=1)
dfCol_Ok=dfCol_Ok.drop(labels=variables, axis=1)
dfCol_Ok.head() # mostrar que quedo
```

# Cargar datos en la tabla de Redshift

```python
# Crear sesion de Spark

import os
import psycopg2

from pyspark.sql import SparkSession
from pyspark.sql.functions import when, lit, col

# Postgres and Redshift JDBCs
driver_path = "/home/working_dir/Entregables/driver_jdbc/postgresql-42.2.27.jre7.jar"

os.environ['PYSPARK_SUBMIT_ARGS'] = f'--driver-class-path {driver_path} --jars {driver_path} pyspark-shell'
os.environ['SPARK_CLASSPATH'] = driver_path

# Create SparkSession
spark = SparkSession.builder \
        .master("local") \
        .appName("Conexion entre Pyspark y Redshift") \
        .config("spark.jars", driver_path) \
        .config("spark.executor.extraClassPath", driver_path) \
        .getOrCreate()
```

```python
env = os.environ
```

```python
# Connect to Redshift using psycopg2
conn = psycopg2.connect(
    host=env['AWS_REDSHIFT_HOST'],
    port=env['AWS_REDSHIFT_PORT'],
    dbname=env['AWS_REDSHIFT_DBNAME'],
    user=env['AWS_REDSHIFT_USER'],
    password=env['AWS_REDSHIFT_PASSWORD']
)
```

```python
cursor = conn.cursor()
cursor.execute(f"""
create table if not exists {env['AWS_REDSHIFT_SCHEMA']}.dfCol (
    "1" INT,
    "2" INT,
    "F" INT,
    "M" INT,
    "Comunitaria" INT,
    "Relacionado" INT,
    "fecha_reporte_web" DATETIME,
    "id_de_caso" INT distkey,
    "fecha_de_notificaci_n" DATETIME,
    "departamento" INT,
    "ciudad_municipio" INT,
    "edad" INT,
    "fecha_inicio_sintomas" DATETIME,
    "fecha_diagnostico" DATETIME,
    "per_etn_" INT
) sortkey(id_de_caso);
""")
conn.commit()
cursor.close()
print("Table created!")
```

```python
cursor = conn.cursor()
cursor.execute(f"""
SELECT
  distinct tablename
FROM
  PG_TABLE_DEF
WHERE
  schemaname = '{env['AWS_REDSHIFT_SCHEMA']}';
""")
# resultado = cursor.fetchall()
print(", ".join(map(lambda x: x[0], cursor.fetchall())))
cursor.close()
```

```python
# Create the DataFrame with the specified column names
dfCol3 = spark.createDataFrame(dfCol_Ok['data'], ["1", "2", "F", "M", "Comunitaria", "Relacionado", "fecha_reporte_web", "id_de_caso", "fecha_de_notificaci_n", "departamento", "ciudad_municipio", "edad", "fecha_inicio_sintomas", "fecha_diagnostico", "per_etn_" ])
```

```python
dfCol3.printSchema()
dfCol3.show()
```

```python
dfCol_to_write = dfCol3.withColumn('frequency', lit('Mensual'))
dfCol_to_write.printSchema()
dfCol_to_write.show()
```

```python
dfCol_to_write.write \
    .format("jdbc") \
    .option("url", f"jdbc:postgresql://{env['AWS_REDSHIFT_HOST']}:{env['AWS_REDSHIFT_PORT']}/{env['AWS_REDSHIFT_DBNAME']}") \
    .option("dbtable", f"{env['AWS_REDSHIFT_SCHEMA']}.exportaciones_cereales") \
    .option("user", env['AWS_REDSHIFT_USER']) \
    .option("password", env['AWS_REDSHIFT_PASSWORD']) \
    .option("driver", "org.postgresql.Driver") \
    .mode("overwrite") \
    .save()
```

```python
# Query Redshift using Spark SQL
query = f"select * from {env['AWS_REDSHIFT_SCHEMA']}.dfCol"
data = spark.read \
    .format("jdbc") \
    .option("url", f"jdbc:postgresql://{env['AWS_REDSHIFT_HOST']}:{env['AWS_REDSHIFT_PORT']}/{env['AWS_REDSHIFT_DBNAME']}") \
    .option("dbtable", f"({query}) as tmp_table") \
    .option("user", env['AWS_REDSHIFT_USER']) \
    .option("password", env['AWS_REDSHIFT_PASSWORD']) \
    .option("driver", "org.postgresql.Driver") \
    .load()
```
